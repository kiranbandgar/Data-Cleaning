```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/train (1).csv')
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(data.isnull())
```

```python
null_var=data.isnull().sum()/data.shape[0]*100
```

```python
drop_col=null_var[null_var >20].keys()
drop_col
```

```python
data_drop=data.drop(columns=drop_col)
```

```python
data_drop.head()
```

```python
sns.heatmap(data_drop.isnull())
```

```python
categorey=['LotFrontage','GarageYrBlt']
for feature in categorey:
  data_drop[feature]=data_drop[feature].fillna(data_drop[feature].median())
```

```python
categorey1=['MasVnrType','MasVnrArea','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinSF1','BsmtFinType2','Electrical','GarageType','GarageFinish','GarageQual','GarageCond']
for feature in categorey1:
  data_drop[feature]=data_drop[feature].fillna(data_drop[feature].mode()[0])
```
